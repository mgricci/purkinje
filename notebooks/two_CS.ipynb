```python
from purkinje.models import Purkinje
from purkinje.utils import ensure_dir
from purkinje.visualize import plot_raster
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import os
import numpy as np
```

# Simulation 5, part 1: Two CSs

We show how only the longest CS is learned when two CSs are supplied per trial. 

## Set parameters

```python
acq_trials=400
CS_onsets=[0.0, .3] # Control CS timings here. 
CS_offsets=[.1, .4]
US_onsets=[.5]
US_offsets=[.6]
probe_start = 50
probe_per   = 5
ITI=10
CS_freq=50
US_freq = 500
```

## Train

```python
pk = Purkinje(a=4000)
pk.run(CS_onsets=CS_onsets, CS_offsets=CS_offsets, US_onsets=US_onsets, US_offsets=US_offsets,
       ITI=ITI, CS_freq=CS_freq, US_freq=US_freq,trials=acq_trials, probe_start=probe_start,
       probe_per=probe_per,verbose=0)
```

## Visualize

```python
#  Rasters
fig_save_dir = '../figures'
ensure_dir(fig_save_dir)
save_path = os.path.join(fig_save_dir, 'two_cs.pdf')

raster = np.array(pk.all_rasters)

plot_raster(raster, CS_onsets, CS_offsets, US_onsets, probe_start, probe_per, save_path)
```
