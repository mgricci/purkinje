```python
from purkinje.models import Purkinje
from purkinje.utils import ensure_dir
from purkinje.visualize import plot_raster
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import os
import numpy as np
```

# Simulation 5, part 2: Two USs

Here, we show that when two USs are presented per trial, only the first is learned. 

## Set parameters

```python
trials=800
CS_onsets=[0.0]
CS_offsets=[.1]
US_onsets=[.2, .4] # Set US stimulation timing here
US_offsets=[.225, .425]
probe_start = 50
probe_per   = 5
ITI=10
CS_freq=50
US_freq = 500
```

## Train

```python
pk = Purkinje(a=2000)
pk.run(CS_onsets=CS_onsets, CS_offsets=CS_offsets, US_onsets=US_onsets, US_offsets=US_offsets,
       ITI=ITI, CS_freq=CS_freq, US_freq=US_freq,trials=trials, probe_start=probe_start,
       probe_per=probe_per,verbose=0)
```

## Visualize

```python
#  Rasters
fig_save_dir = '../figures'
ensure_dir(fig_save_dir)
save_path = os.path.join(fig_save_dir, 'two_us.pdf')

raster = np.array(pk.all_rasters)

plot_raster(raster, CS_onsets, CS_offsets, US_onsets, probe_start, probe_per, save_path)
```
